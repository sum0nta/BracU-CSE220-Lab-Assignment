```python
class DoublyNode:
    def __init__(self,elem,next = None, prev = None):
        self.elem = elem
        self.next, self.prev = next, prev
class Patient:
    def __init__(self,id,name,age,bg):
        self.id = id
        self.name = name
        self.age = age
        self.bg = bg
class WRM:
    def __init__(self):
        self.head = DoublyNode(0,None,None)
    
    def RegisterPatient(self,id, name, age, bloodgroup):
        h = self.head
        p1 = Patient(id, name, age, bloodgroup)
        if self.head.next == None:
            self.head.next = DoublyNode(p1)
            self.head.next.prev = self.head
        else:
            while h.next.elem != 0:
                h = h.next
            h.next = DoublyNode(p1)
            h.next.prev = h
        h.next.next = self.head
        self.head.prev = h.next
        print(f"{p1.name} has succesfully been added")
    
    def ServePatient(self):
        if self.head.next != None:
            h = self.head.next
            self.head.next = self.head.next.next
            self.head.next.prev = self.head
            h.next, h.prev = None, None
            print(f"{h.elem.name} is being served")
            return h
        else:
            print("No patient available")
    
    def CancelAll(self):
        self.head.next = None
        self.head.prev = None
        print(f"Success")

    def CanDoctorGoHome(self):
        if self.head.next == None:
            print("Yes")
        else:
            print("No")
    
    def ShowAllPatient(self):
        h = self.head.next
        print("All Patients: ",end = "")
        while h.elem != 0:
            print(h.elem.name,end = " ")
            h = h.next
        print()

c1 = WRM()
c1.CanDoctorGoHome()
c1.RegisterPatient(1,"Anne",21,"A+")
c1.RegisterPatient(2,"Bnne",22,"A+")
c1.RegisterPatient(3,"Cnne",23,"A+")
c1.RegisterPatient(4,"Dnne",24,"A+")
c1.ServePatient()
c1.ServePatient()
c1.ServePatient()
c1.ShowAllPatient()
c1.CanDoctorGoHome()
c1.CancelAll()
c1.CanDoctorGoHome()
```
