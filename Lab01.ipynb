```python
# Complete the functions defined in this cell


# Test 01: Play Right

def rightRotate(arr):
  last = arr[len(arr)-1]
  for i in range(len(arr)-1,-1,-1):
    arr[i] = arr[i-1]
  arr[0] = last
  return arr

def playRight(sequence,beats):
  for i in range(len(beats)):
    if beats[i] == 1:
      rightRotate(sequence)
  return sequence


# Test 02: Discard Cards

def leftShift(arr,start,end):
  for i in range(start,end-1):
    arr[i] = arr[i+1]
  arr[end-1] = 0
  return arr

def discardCards(cards,number):
  i = 0
  while i < len(cards):
    if cards[i] == number:
      leftShift(cards,i,len(cards))
    else:
      i += 1
  return cards

#Test 03: Merge Lineup

def reverse(arr):
  for i in range(len(arr)//2):
    arr[i],arr[len(arr)-i-1] = arr[len(arr)-i-1], arr[i]
  return arr

def mergeLineup(pokemon_1, pokemon_2):
  for i in range(len(pokemon_1)):
    if pokemon_1[i] == None:
      pokemon_1[i] = 0
    if pokemon_2[i] == None:
      pokemon_2[i] = 0
  pokemon_2 = reverse(pokemon_2)
 
  arr = np.zeros(len(pokemon_1),dtype = int)
  for i in range(len(pokemon_1)):
    arr[i] = pokemon_1[i] + pokemon_2[i]

  return arr

# Test 04: Balance your Salami
def sum(arr):
    total = 0
    for i in range(len(arr)):
        total += arr[i]
    return total
def balanceSalami(salami):
    total_sum = sum(salami)
    running_sum = 0

    for i in range(len(salami) - 1):
        running_sum += salami[i]
        if running_sum == total_sum / 2:
            return True

    return False

# Test 05: Protecc Salami
def insert(arr,val):
    for i in range(len(arr)):
        if arr[i] == 0:
            arr[i] = val
            break
    return arr

def protectSalami(salami):
    arr_nums = np.zeros(len(salami),dtype=int)
    arr_occurences = np.zeros(len(salami),dtype=int)
    for i in range(len(salami)):
        total = 0
        if salami[i]!= 0 and salami[i] not in arr_nums:
            arr_nums = insert(arr_nums,salami[i])

            for j in range(i,len(salami)):
                if salami[i] == salami[j]:
                    total += 1
        arr_occurences = insert(arr_occurences,total)
    
    checking_arr = np.zeros(len(salami))
    for i in range(len(arr_occurences)):
        if arr_occurences[i] > 1 and arr_occurences[i] in checking_arr:
            return True
        checking_arr = insert(checking_arr,arr_occurences[i])          
    return False

# Test 06: Odd Even Wave
def checkAdjacent(x,y):
    if x %2 == 0 and y%2 ==1 :
        return True
    if x %2 == 1 and y%2 ==0 :
        return True
    return False

def waveYourFlag(arr):
    size = len(arr) - 1
    i = 0
    j = 2
    while i < size:
        if checkAdjacent(arr[i],arr[i+1]) == False:
            arr[j],arr[i+1] = arr[i+1],arr[j]
            j += 1
        else:
            i += 2
            j = i + 1
            
    return arr
```
